## Channels : Per Category 📦

    Now we will analysis how many channels there are in each category.

```python
channels_category_df = trend_df[['category_id','channel_title']].copy()
```

```python
channels_category_df=channels_category_df.drop_duplicates('channel_title')
```

## Ratings Disabled : Per channel 😶

    We will explore which channels had the most number of ratings disabled.

```python
rdisabled_per_ch = trend_df.ratings_disabled.copy().to_frame()
```

```python
rdisabled_per_ch['channel_title']= trend_df.channel_title.copy()
```

```python
count_rdisabled = rdisabled_per_ch[rdisabled_per_ch.ratings_disabled == True]
count_rdisabled=count_rdisabled.groupby('channel_title').count()
count_rdisabled=count_rdisabled.sort_values(by='ratings_disabled' , ascending=False)
```

```python
count_rdisabled
```

```python
#to ignore the font missing warnings as some channel titles contains unique fonts 
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(16,20))
plt.title("Ratings disabled per channel")
sns.barplot(x=count_rdisabled.ratings_disabled, y=count_rdisabled.index).set(ylabel='channel name',xlabel='Number of videos on which ratings disabled')
```

As we can see 'only ias' channel had the most number of videos in which ratings were disabled.

### Q2: Which  20 trending videos had the most views ?

```python
sample_df = trend_df.copy()
#dropping all columns except title and views
sample_df.drop(sample_df.columns.difference(['title','views']), 1, inplace=True)
```

```python
#sorted views and dropped all duplicate titles
sample_df=sample_df.sort_values(by='views',ascending=False).drop_duplicates(subset='title').head(20)
sample_df
```

```python
plt.figure(figsize=(12,12))
plt.title('Top 20 Trending videos with most views')
sns.barplot(x=sample_df.views,y=sample_df.title, palette='Spectral').set(ylabel='Video title', xlabel='Views(per 10 crores)')
```

As we can see youtube rewind 2017 was the most viewed trending video in this dataset. YouTube Rewind is a video series produced and created by YouTube and Portal A Interactive. These videos are an overview and a recap of each year's viral videos, events, trends, and music.

### Q3: Which top 20 channels had the most number of comments on the trending videos?

```python
sample_df=trend_df.copy()
#dropping all columns except comments and channel title
sample_df.drop(sample_df.columns.difference(['channel_title','comment_count']),1, inplace=True)
#taking a sum of all comments per channel
sample_df=sample_df.groupby('channel_title').sum().sort_values(by='comment_count', ascending=False).head(20)
```

```python
sample_df
```

```python
plt.figure(figsize=(12,8))
plt.title('Most number of comments per channel')
plt.xticks(rotation=75)
sns.barplot(x=sample_df.index, y=sample_df.comment_count, palette='mako').set(ylabel='comment count(Per 10 lakh)',xlabel='Channel title')
```

As we can see Technical Guruji had the most number of comments in his trending videos.😂

### Q4: Which top 20 channels had the most number of comments disabled on their trending videos?

```python
sample_df = trend_df.copy()
```

```python
sample_df.drop(sample_df.columns.difference(['channel_title', 'comments_disabled']),1,inplace=True )
```

```python
sample_df=sample_df[sample_df.comments_disabled == True]
#taking the number of comments disabled per channel
sample_df=sample_df.groupby('channel_title').sum().sort_values(by='comments_disabled', ascending=False).head(20)
sample_df
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=75)
plt.title('Top 20 channels that had their comments section disabled in trending videos')
sns.barplot(x=sample_df.index, y=sample_df.comments_disabled).set(xlabel='Channel Name', ylabel='Number of videos on which comments section disabled')
```

Again as we see only ias had their comments section disabled as well their ratings.
